```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
def plot_data(file_name,title):
    y = []
    with open(file_name) as f:
        for line in f:
            temp = line.rstrip(os.linesep)
            y.append(float(temp))
        f.close()
    ax = sns.distplot(y)
    ax.set_xlabel('logBB')
    ax.set_ylabel('Probability Density')
    ax.set_title(title)
    return y
```

```python
y1 = plot_data('../data/y_train.txt','y_train logBB Distribution')
```

```python
y2 = plot_data('../data/y_test.txt','y_test logBB Distribution')
```

```python
ax = sns.distplot(y1,hist=False,label='y_train')
sns.distplot(y2,hist=False,label='y_test')
ax.set_xlabel('logBB')
ax.set_ylabel('Probability Density')
ax.set_title('logBB Distribution')
plt.legend(loc='upper left')
```
